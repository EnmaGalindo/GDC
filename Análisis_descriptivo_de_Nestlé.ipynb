```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_df = pd.read_csv('Nestle__Climate_Action_Database.csv')
data_df
```

Estadísticas descriptivas básicas

```python
print(data_df.columns)
```

```python
print(data_df.describe())
```

Análisis de tendencias a lo largo del tiempo (Series de tiempo)

```python
# Gráfico de series de tiempo para ver la evolución de las variables a lo largo de los años
plt.figure(figsize=(10,6))

# Graficamos la reducción de emisiones, energía renovable y productos basados en plantas
plt.plot(data_df['year'], data_df['ghg_emissions_reduction'], label='Reducción de Emisiones de GEI (%)', color='green')
plt.plot(data_df['year'], data_df['renewable_energy_percentage'], label='Energía Renovable (%)', color='blue')
plt.plot(data_df['year'], data_df['plant_based_products'], label='Productos Basados en Plantas (%)', color='orange')

# Añadir título y etiquetas
plt.title('Evolución de Variables Climáticas de Nestlé a lo Largo del Tiempo')
plt.xlabel('Año')
plt.ylabel('Valor')
plt.legend()
plt.grid(True)
plt.show()
```

Histogramas de las Variables

```python
# Crear histogramas individuales para cada variable clave

# Graficar el histograma para la reducción de emisiones de GEI
plt.figure(figsize=(10,6))
plt.hist(data_df['ghg_emissions_reduction'], bins=20, color='green', alpha=0.7)
plt.title('Distribución de Reducción de Emisiones de GEI')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Graficar el histograma para la energía renovable
plt.figure(figsize=(10,6))
plt.hist(data_df['renewable_energy_percentage'], bins=20, color='blue', alpha=0.7)
plt.title('Distribución de Energía Renovable')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# Graficar el histograma para productos basados en plantas
plt.figure(figsize=(10,6))
plt.hist(data_df['plant_based_products'], bins=20, color='orange', alpha=0.7)
plt.title('Distribución de Productos Basados en Plantas')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

Gráficos de Cajas y Bigotes (Boxplots)

```python
# Gráfico de cajas para las variables clave
plt.figure(figsize=(10,6))

# Boxplot para la reducción de emisiones
plt.subplot(1, 3, 1)
plt.boxplot(data_df['ghg_emissions_reduction'], vert=False, patch_artist=True, boxprops=dict(facecolor='green'))
plt.title('Reducción de Emisiones de GEI')

# Boxplot para la energía renovable
plt.subplot(1, 3, 2)
plt.boxplot(data_df['renewable_energy_percentage'], vert=False, patch_artist=True, boxprops=dict(facecolor='blue'))
plt.title('Energía Renovable')

# Boxplot para productos basados en plantas
plt.subplot(1, 3, 3)
plt.boxplot(data_df['plant_based_products'], vert=False, patch_artist=True, boxprops=dict(facecolor='orange'))
plt.title('Productos Basados en Plantas')

# Ajustar la disposición de los subgráficos
plt.tight_layout()
plt.show()
```

Cálculo de Parámetros Estadísticos Clave

```python
# Calcular la media, desviación estándar y varianza de las variables
mean_values = data_df.mean()
std_dev_values = data_df.std()
variance_values = data_df.var()

# Mostrar los resultados
print("Media de las variables:")
print(mean_values)
print("\nDesviación estándar de las variables:")
print(std_dev_values)
print("\nVarianza de las variables:")
print(variance_values)
```

Correlación entre Variables

```python
# Calcular la matriz de correlación
correlation_matrix = data_df.corr()

# Crear el mapa de calor de la correlación
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matriz de Correlación entre Variables Climáticas")
plt.show()
```

Identificación de Outliers (Valores Atípicos)

```python
# Calcular el rango intercuartílico (IQR) para detectar outliers
Q1 = data_df.quantile(0.25)
Q3 = data_df.quantile(0.75)
IQR = Q3 - Q1

# Identificar outliers
outliers = (data_df < (Q1 - 1.5 * IQR)) | (data_df > (Q3 + 1.5 * IQR))

# Mostrar los valores atípicos
print("Valores atípicos detectados en las variables:")
print(outliers)
```
